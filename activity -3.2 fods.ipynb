```python
import numpy as np
import pandas as pd
```

```python
prices = np.array([2.5, 3.0, 2.8, 3.2, 15.0, 2.9, 3.1, 2.7, 100.0])
print("Original Data:", prices)
```

```python
from scipy.stats import zscore

z_scores = zscore(prices)
print("Z-Scores:", z_scores)

outliers = prices[np.abs(z_scores) > 3]
print("Detected Outliers:", outliers)
```

```python
prices_corrected = prices.copy()

# Replace 100.0 with 10.0
prices_corrected[prices_corrected == 100.0] = 10.0

print("After Correcting Typo:", prices_corrected)
```

```python
z_scores_new = zscore(prices_corrected)
outliers_new = prices_corrected[np.abs(z_scores_new) > 3]

print("New Z-Scores:", z_scores_new)
print("Outliers After Correction:", outliers_new)
```

```python
Q1 = np.percentile(prices_corrected, 25)
Q3 = np.percentile(prices_corrected, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

iqr_outliers = prices_corrected[(prices_corrected < lower_bound) |
                                (prices_corrected > upper_bound)]

print("Outliers using IQR:", iqr_outliers)
```

```python
prices_capped = np.clip(prices_corrected, None, upper_bound)
print("After Capping:", prices_capped)
```

```python
log_prices = np.log(prices_corrected)
print("Log Transformed Data:", log_prices)
```
