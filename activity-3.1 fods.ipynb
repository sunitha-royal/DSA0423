```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, zscore

# For reproducibility
np.random.seed(42)

# Generate datasets
normal_data = np.random.normal(loc=50, scale=5, size=1000)
skewed_data = np.random.exponential(scale=10, size=1000)
```

```python
plt.figure()
sns.histplot(normal_data, kde=True)
plt.title("Normal Distribution")
plt.show()

plt.figure()
sns.histplot(skewed_data, kde=True)
plt.title("Skewed (Exponential) Distribution")
plt.show()
```

```python
print("Normal Data Skewness:", skew(normal_data))
print("Skewed Data Skewness:", skew(skewed_data))

print("Normal Data Kurtosis (Fisher):", kurtosis(normal_data))
print("Skewed Data Kurtosis (Fisher):", kurtosis(skewed_data))
```

```python
def detect_outliers(data):
    z_scores = zscore(data)
    outliers = np.where(np.abs(z_scores) > 3)
    return len(outliers[0])

outlier_count = detect_outliers(skewed_data)
print("Number of outliers in skewed data:", outlier_count)
```
